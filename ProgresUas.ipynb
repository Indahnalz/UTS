## Memvisualisasikan data

```python
# pengaturan visualisasi

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# jumlah kelas dalam kolom target 
kepuasan_konsumen["satisfaction"].value_counts()
```

```python
# Memvisualisasikan distribusi kepuasan di antara para penumpang
fig, (ax1, ax2) = plt.subplots(nrows=1,ncols=2, figsize=(10,3))
ax1.bar(kepuasan_konsumen["satisfaction"].value_counts().index, kepuasan_konsumen["satisfaction"].value_counts(),color=["red", "orange"])
ax1.set(title="Satisfaction Distribution",ylabel="Count")
ax2.pie(kepuasan_konsumen["satisfaction"].value_counts(),colors=["red", "orange"],
        labels= ["Neutral or Dissatisfied", "Satisfied"] , 
        autopct='%1.1f%%',explode=[0.05,0.05] ,startangle=0, shadow = True)
ax2.set(title="Satisfaction Distribution");
```
